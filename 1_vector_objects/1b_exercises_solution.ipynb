### Write the folowing functions
They should work on arbitrary Polygons

```python
# pages 9-13
def area_net(polygon):
    val = polygon.area
    return val

def area_filled(polygon):
    val = Polygon(polygon.exterior).area
    return val

def area_convex(polygon):
    val = polygon.convex_hull.area
    return val

def perimeter(polygon):
    val = polygon.exterior.length
    return val

def perimeter_convex(polygon):
    val = polygon.convex_hull.length
    return val


# page 21
def compactness(polygon):
    val = 4*pi*polygon.area/(perimeter(polygon))**2
    return val

# page 30
def circularity(polygon):
    val = 4*pi*polygon.area/(perimeter_convex(polygon))**2
    return val


# page 33
def convexity(polygon):
    val = perimeter_convex(polygon)/perimeter(polygon)
    return val

# page 41
def solidity(polygon):
    val = area_net(polygon)/area_convex(polygon)
    return val

# page 23
def elongation(polygon):
    pointA= Point(triangle.minimum_rotated_rectangle.boundary.coords[0])
    point0=Point(triangle.minimum_rotated_rectangle.boundary.coords[1])
    pointB=Point(triangle.minimum_rotated_rectangle.boundary.coords[2])
    el = point0.distance(pointA)/point0.distance(pointB)
    val = el if el<1 else 1/el
    return val

# page 52
def euler_number(polygon):
    val = 1 - len(polygon2.interiors)
    return val
```

```python
def print_descriptors(polygon):
    display(polygon)
    print("Area (net): ",area_net(polygon))
    print("Area (filled): ",area_filled(polygon))
    print("Area (convex): ",area_convex(polygon))
    print("Perimeter: ",perimeter(polygon))
    print("Perimeter (convex): ",perimeter_convex(polygon))
    print("Compactness: ",compactness(polygon))
    print("Circularity: ",circularity(polygon))
    print("Convexity: ",convexity(polygon))
    print("Solidity: ",solidity(polygon))
    print("Elongation: ",elongation(polygon))
    print("Euler number: ",euler_number(polygon))
```
